```python
import numpy as np
import matplotlib.pyplot as plt
aa = np.linspace(-100,100,201)
print(aa)
```

```python
print ( aa.size )
```

```python
bb = np.tan(aa)
print(bb)
```

```python
print ( bb.size )
```

```python
plt.plot (aa,bb)
plt.grid()
plt.show()
```

```python
cc = aa*aa*aa
print(cc)
```

```python
plt.plot (aa,cc)
plt.grid()
plt.show()
```

```python
dd = aa*aa + aa*6 + 4  
print(dd)
```

```python
plt.plot (aa,dd)
plt.grid()
plt.show()
```
